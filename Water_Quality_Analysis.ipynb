## Decision Tree

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("WaterQuality.csv")

# Display the first few rows of the dataframe
df.head()

# Display the information about the dataframe
df.info()

# Data cleaning
df.replace("#NUM!", float("nan"), inplace=True)
df = df.dropna()

# Splitting the data into features (X) and target (y)
X = df.drop(["is_safe"], axis="columns")
y = df.is_safe.astype(int)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=10)

# Initializing and training the decision tree classifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

# Evaluating the model
print("Model Accuracy:", dt.score(X_test, y_test))

# Plotting the decision tree
plt.figure(figsize=(20,10))
tree.plot_tree(dt, feature_names=X.columns, class_names=["Not Safe", "Safe"], filled=True)
plt.show()
```

## K Nearest Neighbors

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

# Load the data
df = pd.read_csv('WaterQuality.csv')

# Display the first few rows of the dataframe
print(df.head())

# Display the information about the dataframe
print(df.info())

# Data cleaning
df.replace('#NUM!', float('nan'), inplace=True)
df = df.dropna()

# Splitting the data into features (X) and target (y)
X = df.drop(["is_safe"], axis="columns")
y = df.is_safe.astype(int)

# Standardizing the features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=12)

# Initializing and training the K Nearest Neighbors classifier
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train, y_train)

# Evaluating the model
print("Model Accuracy:", knn.score(X_test, y_test))
```

## Logistic Regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Assuming you've loaded the dataset and preprocessed it as per your earlier steps
df = pd.read_csv("WaterQuality.csv")
df.replace("#NUM!", float("nan"), inplace=True)
df = df.dropna()

# Splitting data into features (x) and target (y)
x = df.drop("is_safe", axis=1)
y = df["is_safe"]

# Splitting into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=12)

# Initializing Logistic Regression model and fitting on training data
lr = LogisticRegression(max_iter=1000)
lr.fit(x_train, y_train)

# Evaluating model accuracy on test data
accuracy = lr.score(x_test, y_test)
print("Logistic Regression Accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Plotting confusion matrix
y_pred = lr.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Unsafe', 'Safe'])
plt.figure(figsize=(8, 6))
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()

```

## Multi Layer Perceptron (MLP)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
df = pd.read_csv("WaterQuality.csv")
df.head()
```

```python
df.replace("#NUM!", float("nan"), inplace=True)
df = df.dropna()
```

```python
x = df.drop("is_safe", axis=1)
y = df["is_safe"]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=12)

mlp = MLPClassifier(hidden_layer_sizes=(200), max_iter=1000)
mlp.fit(x_train, y_train)
mlp.score(x_train, y_train)
```
